## 概览OOP

### 属性继承搜索

```python
object.attribute
```
python中通过上面的方式调用对象的方法，使用时会启动搜索，搜索对象连接的类树，来寻找`attribute`首次出现的类。首先搜索`object`，然后是该对象之上的所有类，由下往上，由左到右。

树中位置较低的类称为子类，较高的是父类，父类提供了所有子类共享行为，因为搜索过程是自底向上的，所以子类可能会在较低的位置重新定义父类的名称，从而覆盖父类定义的行为。

### 类和实例

- 类：类是实例工厂，类的属性提供了行为，所有从类产生的实例都继承该类的属性。
- 实例：代表程序领域中具体的元素，记录了每个实例自己的数据。

### 方法调用

如果`I`是`C`的实例，那么`I.w()`调用映射为`C.w(I)`，python把隐含的实例传入方法中的第一位参数，习惯上将其称为`self`。

### 类树

- 每个class语句会生成一个新的类方法
- 每次类调用时，就会产生一个新的实例对象
- 实例自动链接到创建它们的类
- 类链接到其父类的方式是，将父类列在class头部的括号内；括号从左至右的顺序决定树中的次序。

python可以多继承，就是有多个父类；附加在实例上的属性只属于那些实例，但附加在类上的属性则由所有子类和实例共享。

### 类和多态

通过编写子类定制每个特殊类型的不同行为，其他共同行为则有父类定义。多态就是运算取决子子类的具体实现。
