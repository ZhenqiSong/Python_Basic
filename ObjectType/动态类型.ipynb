- 变量在第一次赋值时创建
- 变量类型不是静态的，同一个变量可以赋值任意类型对象
- 再使用变量前要保证已经赋值

```python
a = 3
```

上面的赋值语句执行了三个步骤：

- 创建一个整数对象，值为3
- 如果a首次使用，创建变量a
- 链接变量a和对象3

变量和对象互相独立，只是关联起来，可以用变量调用对象的值，称为**引用**  
每个对象有两个头：一个表示该对象的类型，一个计数器表示引用量

python的垃圾回收机制，当一个对象的应用数为0时，将回收该队形的内存，销毁该对象。

## 共享引用

```python
a = 3
b = a
```

```python
id(a), id(b)
```

```python
a += 2
id(a), id(b)
```

```python
L = [1, 2, 3]
M = L
```

```python
id(M), id(L)
```

```python
L[0] = 4
M # 说明共享同一个对象
```

```python
M = L.copy() # 使用copy进行深拷贝
L[0] = 1
M
```

```python
L = [1, [2, 3], 4]
M = L.copy() # 使用copy直接进行拷贝，里面的list仍然是同一个
L[1][0] = 5
L[0] = 5
M 
```

```python
import copy
```

```python
M = copy.deepcopy(L)
print(L, M)
L[1][0] = 2
print(L, M)
```

```python
L = [1, 2, 3]
M = [1, 2, 3]
print(L==M, L is M) # == 比较的是值 is比较的是对象
```

```python
import sys
L = 42 # 这是因为整数等类型引用为0时，并没有被直接回收，而是再缓存中，当再次使用时，直接用缓存中的值。
print(id(L))
L = 50
M = 42
print(id(M))
print(sys.getrefcount(42)) # 引用的数量
print(sys.getrefcount([]))
```
