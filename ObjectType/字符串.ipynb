## 字符串字面值

```python
print('szq')
print("sz'q")
a = """
    szq
    """
print(a)
print("s\tp\n")
print(r"\a")
```

- 字符串是不可变的有序序列
- 单引号双引号作用相同
- 使用r""可以屏蔽转义符号的作用，所见即所得
- 使用三个引号可以表示多行代码块, 同样屏蔽转义

### 索引和切片

```python
a = "python"
```

```python
print(a[0], a[-2]) # -2等同于len(a)-2
```

```python
a[:]
```

```python
a[1:3] # 切片从1开始到3结束，而不包括3，所以要想取到最后，不能用[:-1],而是用[:]
```

使用切片可以实现深拷贝

切分支持三个参数的操作，增加一个步长

```python
a = "0123456789"
```

```python
a[0:8:2] # 最后一个值是步长，可以根据步长间隔取值
```

```python
a[8:1:-2] # 逆序取间隔，必须第一个值大于第二个
```

### 改变字符串的值

```python
# 字符串是不可修改的，但是通过重新
```

```python
a = "hello, word"
a = a[:-4] + "world"
print(a)
```

```python
c = a.replace("world", "word")
print(c)
```

```python
where = c.find("word")
print(where)
```

### 格式化字符串
![](D:/SongZhenqi/Project/Python/Python_Basic/images/stringformat.png)

```python
# 除了类似c风格的字符串表达式，还有字典型的
value = {"name":"szq", "age":18}
"%(name)s : %(age)d"%value
```

```python
import sys 
"My {1[kind]} runs {0.platform}".format(sys, {"kind":"laptop"})
```

```python
"{0} {1}".format("a", "b")
```

```python
"{name}".format(name="szq")
```

```python
"{0:>10} = {0:<10}".format("szq", 10)
# 冒号后面是格式设置，10表示宽度，> 是右对齐， < 左对齐
```

```python
{"{0:e}, {1:.3e}, {2:g}".format(3.1415926, 3.1425926, 3.1415926)}
```

```python
"{0:X}, {1:o}, {2:b}".format(255, 255, 255)
```
